The following code snippet will do the first task. It will read every single line of the csv document, extract the values by using column names, and print them in a for loop.

```python
import csv
from datetime import datetime

# class Avocado:
#  initiate the class and assign a value on each Avocado variable
        
#     def printer():
#         print each variable
if __name__ == '__main__':

    with open('/content/avocado.csv') as csvfile:
        readCSV = csv.DictReader(csvfile, delimiter=',')
        avocadosales = []
        for col in readCSV:
            print(col['Date'],col['AveragePrice'], col['Total Volume'], col['Total Bags'],col['Small Bags'],col['Large Bags'],col['XLarge Bags'], col['year'],col['region'] )
```

Now it is time to create a class for Avocado prices!

```python
class Avocado:
  def __init__(self, date, averagePrice, totalVolume, totalBags, smallBags, largeBags, xLargeBags, year, region):
      self.date = date
      self.averagePrice = averagePrice
      self.totalVolume = totalVolume
      self.totalBags = totalBags
      self.smallBags = smallBags
      self.largeBags = largeBags
      self.xLargeBags = xLargeBags
      self.year = year
      self.region = region
  def printer(self):
      print (self.date, self.averagePrice, self.totalBags, self.totalBags, self.smallBags, self.largeBags, self.xLargeBags, self.year, self.region)

```

```python
avocadosales = []
with open('/content/avocado.csv') as csvfile:
        readCSV = csv.DictReader(csvfile, delimiter=',')
        for col in readCSV:
          salesrow = Avocado(col['Date'],col['AveragePrice'], col['Total Volume'], col['Total Bags'],col['Small Bags'],col['Large Bags'],col['XLarge Bags'], col['year'],col['region'])
          avocadosales.append(salesrow)
```

```python
for element in avocadosales:
    print (element.date, element.averagePrice, element.totalBags, element.totalBags, element.smallBags, element.largeBags, element.xLargeBags, element.year, element.region)
```

```python
for element in avocadosales:
    element.printer()
```
